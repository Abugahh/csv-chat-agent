## First using langchain csv agent

This code is setting up a CSV agent using LangChain's experimental toolkit with Google's Generative AI (Gemini Pro). langchain: The core LangChain framework for building LLM applications
langchain_experimental: Contains experimental features not yet in the main LangChain package
langchain-google-genai: LangChain integration with Google's Generative AI models

Creates a connection to Google's Gemini Pro model
Creates a "CSV agent" that can interact with the specified CSV file

How the CSV Agent Works:

Agent Architecture: The CSV agent is a specialized LangChain agent that can reason about and interact with CSV data. It combines:

An LLM (Gemini Pro in this case)
A set of tools for working with CSV files
A reasoning system to determine which tools to use

The agent uses what LangChain calls "ReAct" (Reasoning and Acting)
It thinks about what to do
Takes actions
Observes results
Repeats until it has an answer

Verbose Mode: Setting verbose=True means it will show you its reasoning process, including:

The thoughts it has
The tools it selects
The actions it takes
The observations it makes

```python
from langchain_experimental.agents.agent_toolkits import create_csv_agent
from langchain_google_genai import ChatGoogleGenerativeAI

from langchain.memory import ConversationBufferMemory
from langchain.tools import tool
llm = ChatGoogleGenerativeAI(model="gemini-1.5-pro-001")

memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)


# Create the agent with memory and custom tools
agent = create_csv_agent(
    llm,
    '/content/GCDofDisease_Metaanalysis_Database.csv',
    verbose=True,
    allow_dangerous_code=True,
    memory=memory
)
```
